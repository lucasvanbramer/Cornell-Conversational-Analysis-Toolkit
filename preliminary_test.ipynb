```
import polyglot
from polyglot.text import Detector
```

```
detector=Detector("Something beyond Biddle, please, 中文测试 since he does recognize the validity of Jackson's position regarding Nimitz, even though it he argues that it opened the trials up to ''tu quoque'' criticism. ")
print(len(detector.languages))
for language in detector.languages:
  print(language.confidence)
```

```
count=0
track=''
for utt in wikiconv_2003.iter_utterances():
  text = ''.join(x for x in utt.text if x.isprintable())
  if len(text) == 0 or text == ' ':
    continue
  track=text
  detector=Detector(text, quiet=True)
  languages=[language.name for language in detector.languages if language.confidence > 0]
  if len(languages) > 1:
    print(text)
    print(languages)
    count+=1
print(count)
```

```
print(count)
print(len(wikiconv_2003.utterances))
print(str(count/len(wikiconv_2003.utterances)*100)+"%")
```
